```python
def scrape_next_month():
    for i in range(1,33):
        time.sleep(6)
        #check if there are any archives, if yes return, no return "No Archive"
        try:
            driver.find_element_by_xpath("//ol/article")
            header = driver.find_element_by_xpath("//ol[@class='WSJTheme__list-reset_2LMJV_G3ewWx1L0AiKTiZ9 ']")
            empty.append(header.text)
        except:
            pass
            empty.append("No Archive")
        finally:
            print(len(empty))
            driver.find_element_by_xpath("//div[@class='DayPickerInput']").click()
            try:
                result=driver.find_element_by_xpath("//div[@class='DayPicker-Day'][.='%s']"%(i))
                result.click()
            except:
                pass
                footer = driver.find_element_by_xpath("//span[@class='DayPicker-NavButton DayPicker-NavButton--next']")
                footer.click()
                scrape_next_month()               
```
